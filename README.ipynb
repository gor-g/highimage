```python
import himage as hi

# read an image
img = hi.imread('image.jpg')

# show the image
hi.imshow(img)
```

```python
print(img.shape)
print(img.dtype)
print(img.min(), img.max())

# show different shades of the image
images = [img/i for i in range(1, 10)]
hi.multimshow(images, n_cols=3)
```

```python
# show more shades of the image, with some titles
images = [hi.clip(img*0.3*i) for i in range(1, 10)]
titles = ["shade %d"%i for i in range(1,10)]
hi.multimshow( images, titles, n_cols=4)
```
