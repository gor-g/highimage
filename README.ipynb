```python
import himage as hi

# read an image
img = hi.imread('image.jpg')

# show the image
hi.imshow(img)
```

```python
import numpy as np
print(img.shape)
print(img.dtype)
print(img.min(), img.max())

# show different shades of the image
hi.multimshow([img/i for i in range(1, 10)], n_cols=3)
```

```python
# show more shades of the image, with some titles
hi.multimshow([hi.clip(img*0.3*i) for i in range(1, 10)], 
              ["shade %d"%i for i in range(1,10)],
                n_cols=5)
```
